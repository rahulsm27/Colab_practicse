```python

import numpy as np
from scipy.stats import beta, binom
import matplotlib.pyplot as plt

# Assume we've observed 3 click-throughs out of 5 opened emails
click_throughs = 3
emails_opened = 5

# Set the range of possible click-through rates
xs = np.linspace(0, 1, num=100)

prior_mean = 0.03
belief_strength = 0.5
alpha_prior = prior_mean * belief_strength * 100
beta_prior = (1 - prior_mean) * belief_strength * 100
prior_density = beta.pdf(xs, alpha_prior, beta_prior)

# Calculate the likelihood
likelihood = binom.pmf(click_throughs, emails_opened, xs)

for i in range(1,35,5):
# Calculate the posterior parameters
  alpha_posterior = alpha_prior + click_throughs*i
  beta_posterior = beta_prior + emails_opened*i - click_throughs*i
  # Calculate the posterior density
  posterior_density = beta.pdf(xs, alpha_posterior, beta_posterior)
  plt.plot(posterior_density,label='posterior')
  plt.plot(likelihood,label='likelihood')
  plt.plot(prior_density,label='prior')
  plt.legend()
  plt.show()


```
